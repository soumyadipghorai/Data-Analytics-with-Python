A survey found that 65% of all financial consumers were very satisfied with their primary financial institution. Suppose that 25 financial consumers are sampled and if survey result still holds true today, what is the probability that exactly 19 are very satisfied with their primary financial institute?

````python
p = 0.65
# P(x = 19)

# 25C19*P^19*(1-P)^25-19
print(binom.pmf(k = 19, n = 25, p=p))
````

According to the US census Bureau approx 6% of all workers in Jackson, MIssissippi are unemployed. In conducting a random telephone survey in Jackson, what is the prob of getting two or fewer unemployed workers in a sample of 20?

````python
p = 0.06

# p ( X <= 2) = P(X=0) + P(X=1) + P(X=2)
prob = 0
for i in range(3) :
    prob += binom.pmf(k = i, n = 20, p = p)

print('using pmf --> ', prob)

print('using cdf --> ', binom.cdf(k = 2, n = 20, p = p))
````

#### change of binomial distribution with change in sample size 

````python
for i in range(5) :
    sample = np.random.binomial(n = 100, p = 0.05, size = 10 * pow(10, i))
    sns.histplot(sample);
    print('sample size --> ', 10 * pow(10, i))
    plt.show();
````

#### change of binomial distribution with change in probability 

````python
for i in range(11) :
    if i < 10 :
        sample = np.random.binomial(n = 100, p = 0.01 + (0.1*i), size = 100000)
        prob = 0.01 + (0.1*i)
    else :
        sample = np.random.binomial(n = 100, p = 0.99, size = 100000)
        prob = 0.99
    sns.histplot(sample);
    print('prob --> ', prob)
    plt.show();
````
