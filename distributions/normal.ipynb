What is the probability of obtaining a score greater than 700 on a GMAT test that has a mean of 494 and a sd of 100.

```python
# P(X > 700) = 1 - P(X<=700)
value, mu, sigma = 700, 494, 100

print(1 - norm.cdf(value, mu, sigma))
```

```python
mu, sigma = float(input()), float(input())
print(mu, sigma)
print('mu += sigma --> ', norm.cdf(mu + sigma, mu, sigma) - norm.cdf(mu - sigma, mu, sigma))
print('mu += 2sigma --> ', norm.cdf(mu + 2*sigma, mu, sigma) - norm.cdf(mu - 2*sigma, mu, sigma))
print('mu += 3sigma --> ', norm.cdf(mu + 3*sigma, mu, sigma) - norm.cdf(mu - 3*sigma, mu, sigma))
```

#### change of normal distribution with change in sample size 

```python
for i in range(6) :
    sample = np.random.normal(loc = 0, scale = 1, size = 10 * pow(10, i))
    sns.histplot(sample);
    plt.show();
```

#### change of normal distribution with change in scale

```python
for i in range(10) :
    sample = np.random.normal(loc = 0, scale = 1+(i*3.5), size = 100000)
    sns.histplot(sample);
    plt.xlim(-100 , 100)
    plt.ylim(0 , 2500)
    plt.show();
```

#### change of normal distribution with change in mean

```python
for i in range(10) :
    sample = np.random.normal(loc = 0+(i*1.5), scale = 1, size = 100000)
    sns.histplot(sample);
    plt.xlim(-5 , 20)
    plt.ylim(0 , 2500)
    plt.show();
```
