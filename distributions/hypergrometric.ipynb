suppose 18 major companies are in india and 12 are located in blr. if 3 companies are selected randomly what's the probability that one or more of the selected companies are located in blr.

```python
# P(x >= 1) = 1 - P(x<1) ==> 1 - P(x<=0)
pval = hypergeom.sf(0, 18, 3, 12) # sf = 1-cdf
print(pval)
```

```python
1 - hypergeom.pmf(0, 18, 3, 12)
```

#### change of hypergeometric distribution with change in sample size 

```python
for i in range(1, 6) :
    sample = np.random.hypergeometric(
                ngood= 10 * pow(5, i), nbad= 10 * pow(10, i) - 10 * pow(5, i), 
                nsample= 10 * pow(2, i), size = 10 * pow(10, i)
            )
    print('ngood --> ', 10 * pow(5, i), 'nbad --> ', 10 * pow(10, i) - 10 * pow(5, i), 'nsample --> ', 10 * pow(2, i), 'size --> ', 10 * pow(10, i))
    sns.histplot(sample);
    plt.show();
```

#### change of hypergeometric distribution with change in good and bad ratio  

```python
for i in range(10) :
    # N = 1000 + 6000 
    # k = 1000 
    # n = 1000
    sample = np.random.hypergeometric(
                ngood= 1000 - (i*100), nbad= 6000 + (i*100), 
                nsample= 1000, size = 1000
            )
    print('ngood --> ', 1000 - (i*100), 'nbad --> ', 6000 + (i*100), 'nsample --> ', 1000, 'size --> ', 1000)
    print("{:.100f}".format(hypergeom.pmf(0, 1000 - (i*100) + 6000 + (i*100), 1000, 1000 - (i*100))))
    print("{:.100f}".format(hypergeom.pmf(1000, 1000 - (i*100) + 6000 + (i*100), 1000, 1000 - (i*100))))
    sns.histplot(sample);
    plt.xlim(-5 , 180)
    plt.ylim(0 , 150)
    plt.show();
```
