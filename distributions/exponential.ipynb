A manufactoring firm has been involved in quality control for several years. for production process parts are randomly selected and tested. It has been established that a defective part occurs in a pattern that is poisson with mean 1.38 defects every 10mins during production runs. what is the probability that less than 15min will elapse between any two defects

```python
pois_mu = 1.38
exp_mu = 1/pois_mu
exp_mu
```

```python
# 15/20 = 0.75, loc = 0 because y = (x-loc)/scale nd y = x/scale
expon.cdf(0.75, 0, exp_mu) 
```

#### change of hypergeometric distribution with change in sample size 

```python
for i in range(5) :
    sample = np.random.exponential(scale = 1, size = 10 * pow(10, i))
    sns.histplot(sample);
    plt.show();
```

```python
for i in range(5) :
    sample = np.random.exponential(scale = 1 + (i*2), size = 10000)
    sns.histplot(sample, stat='probability');
    plt.xlim(0 , 100)
    plt.ylim(0 , 0.12)
    plt.show();
```
