Suppose bank customers arrive randomly on weekday afternoons at an average of 3.2 customers every 4mins. Whats the probabililty of exactly 5 customers arriving in a 4min interval on a weekday afternoon.

```python
lmda = 3.2
print(poisson.pmf(5, lmda))
```

Suppose bank customers arrive randomly on weekday afternoons at an average of 3.2 customers every 4mins. Whats the probabililty of having more than 7 customers arriving in a 4min interval on a weekday afternoon.

```python
# P(X > 7) = 1 - P(X<=7)
lmda = 3.2
prob = 0
for i in range(8) :
    prob += poisson.pmf(i, lmda)

print(1 - prob)
```

Suppose bank customers arrive randomly on weekday afternoons at an average of 3.2 customers every 4mins. Whats the probabililty of having exactly 10 customers arriving in a 8min interval on a weekday afternoon.

```python
# !wrong ==> prob > 1
prob = 0
for i in range(11) :
    prob += (poisson.pmf(i, lmda) + poisson.pmf(10 - i, lmda))

print(prob)
```

#### change of poisson distribution with change in sample size 

```python
for i in range(6) :
    sample = np.random.poisson(lam = 5, size = 10 * pow(10, i))
    sns.histplot(sample);
    plt.show();
```
